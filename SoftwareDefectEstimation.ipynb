# Software Defect Estimation Using ML Algorithms
This notebook demonstrates the use of multiple machine learning algorithms to predict software defects using the NASA dataset.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from models import random_forest, naive_bayes, svm, mlp, rbf, bagging
from utils import preprocessing
from graph_plot import plot_accuracies
```

```python
X_train, X_test, y_train, y_test = preprocessing.load_and_preprocess('data/nasa_dataset.csv')
```

```python
models = {
    'Random Forest': random_forest.train_random_forest,
    'Naive Bayes': naive_bayes.train_naive_bayes,
    'SVM': svm.train_svm,
    'MLP': mlp.train_mlp,
    'RBF': rbf.train_rbf,
    'Bagging': bagging.train_bagging
}

accuracies = {}
for name, func in models.items():
    model = func(X_train, y_train)
    acc = accuracy_score(y_test, model.predict(X_test))
    accuracies[name] = acc

plot_accuracies(accuracies)
```
